### 2. Find the number of user reviews and the average rating for each book.<a class="anchor" id="2"></a>

```python
query ='''
SELECT 
    books.book_id,
    books.title,
    AVG(rating) AS rating_avg,
    count (distinct  reviews.review_id) AS reviews_CNT
FROM books  
INNER JOIN ratings ON books.book_id = ratings.book_id
INNER JOIN reviews ON books.book_id = reviews.book_id
GROUP BY
    books.book_id    
ORDER BY
    reviews_CNT DESC  
'''
#print(query)

display(pd.io.sql.read_sql(query, con = engine))
```

some aditional check of the data

Lets look at books with more then 3 reviews

```python
query ='''
SELECT 
    books.book_id,
    books.title,
    AVG(rating) AS rating_avg,
    count (distinct  reviews.review_id) AS reviews_CNT
FROM books  
INNER JOIN ratings ON books.book_id = ratings.book_id
INNER JOIN reviews ON books.book_id = reviews.book_id 
GROUP BY
    books.book_id 
HAVING     
    count (distinct  reviews.review_id) > 3
ORDER BY
    reviews_CNT DESC  
LIMIT 50    
'''
#print(query)

display(pd.io.sql.read_sql(query, con = engine))
```

### 3. Identify the publisher that has released the greatest number of books with more than 50 pages.<a class="anchor" id="3"></a>

```python
query ='''
SELECT 
    publishers.publisher,
    books.publisher_id, 
    count(books.publisher_id) AS published_cnt
FROM 
    books
    INNER JOIN publishers ON publishers.publisher_id = books.publisher_id
WHERE
    num_pages > 50
GROUP BY 
    books.publisher_id,
    publishers.publisher
ORDER BY   
    published_cnt DESC
LIMIT 1    
'''
#print(query)

display(pd.io.sql.read_sql(query, con = engine))
```

Conclusions: <br> 
Penguin Books is the publisher that has published the most books with more then 50 pages.

### 4. Identify the author with the highest average book rating .<a class="anchor" id="4"></a> <br>
look only at books with at least 50 ratings

```python
query ='''
SELECT
    books.author_id, 
    books.title,
    authors.author,
    count(ratings.rating_id) AS rate_cnt,
    AVG(ratings.rating) AS rate
FROM 
    books
    INNER JOIN ratings ON ratings.book_id = books.book_id
    INNER JOIN authors ON authors.author_id = books.author_id
    
GROUP BY
    books.author_id,
    books.title,
    authors.author
HAVING     
    count(ratings.rating_id) > 50
ORDER BY
    rate DESC
LIMIT 1    
'''
#print(query)

display(pd.io.sql.read_sql(query, con = engine))
```

Conclusions: <br> 
the author with the highest book rating are actually a team J.K. Rowling & Mary GrandPré .<br> 
For the book: Harry Potter and the Prisoner of Azkaban


### 5. Find the average number of text reviews among users who rated more than 50 books.<a class="anchor" id="5"></a>

did any user rated the same book twice ?

```python
query ='''
SELECT
   COUNT(book_id) AS all,
   COUNT(DISTINCT book_id) AS distinc,
   (COUNT(book_id) - COUNT(DISTINCT book_id)) AS dif
FROM 
    ratings      
GROUP BY
    username
HAVING   
    COUNT(book_id) - COUNT(DISTINCT book_id) > 0
'''
print(query)


display(pd.io.sql.read_sql(query, con = engine))
```

no user has rated the same book twice

```python
query ='''
SELECT       
    AVG(rev_cnt)
FROM 
    (SELECT
       DISTINCT ratings.username,
       COUNT  (DISTINCT reviews.review_id) AS rev_cnt
    FROM 
        ratings   
        INNER JOIN reviews ON reviews.username = ratings.username    
    GROUP BY
        ratings.username
    HAVING 
        COUNT(ratings.book_id) > 50) AS sub_rat    
'''
#print(query)

#display(pd.io.sql.read_sql(query, con = engine))
```

Conclusions: <br>
the average number or text reviews for readers who have rated more then 50 books is 17.45 <br>
This mean that they only review about every 3rd book they rated.

for that list of users find out haw many reviews they have

```python
query ='''
SELECT
    reviews.username,
    count(reviews.book_id) AS rev_cnt
FROM
    (SELECT       
       username,
       count(book_id) AS rat_cnt
    FROM    
        ratings  
    GROUP BY
        username
    HAVING 
        count(book_id) > 50) AS user_list
    INNER JOIN reviews on user_list.username = reviews.username
WHERE
    user_list.username = reviews.username
GROUP BY
    reviews.username
   
'''
#print(query)

display(pd.io.sql.read_sql(query, con = engine))
```

find the average

```python
query ='''
SELECT 
    AVG(rev_cnt)
FROM (
    SELECT 
        count(reviews.book_id) AS REV_cnt
    FROM
        (SELECT       
           username,
           count(book_id) AS rat_cnt
        FROM    
            ratings  
        GROUP BY
            username
        HAVING 
            count(book_id) > 50) AS user_list
        INNER JOIN reviews on user_list.username = reviews.username
    WHERE
        user_list.username = reviews.username
    GROUP BY
        reviews.username) AS rev_top

'''
#print(query)

display(pd.io.sql.read_sql(query, con = engine))
```
